## Chargement fragment de données en mémoire

```python
with open("Colors.txt",'r') as open_file:
    print('Contenu de Colors.txt :\n' + open_file.read())
```

## Chargement progressif des données via un flux (avec une boucle for)

```python
with open("Colors.txt",'r') as open_file:
    for observation in open_file:
        print('Lecture des data : ' + observation)
```

## Ne lis que les lignes pair du fichier

```python
n = 2
with open("Colors.txt",'r') as open_file:
    for j, observation in enumerate(open_file):
        if j % n == 0:
            print('Ligne lue: ' + str(j)+ ' contenu: ' + observation)
```

## Lecture de fichiers Binaires

```python
# import de la bibliothéque scikit-image
from skimage.io import imread
from skimage.transform import resize
from matplotlib import pyplot as plt
import matplotlib.cm as cm

fic_exemple = ("http://upload.wikimedia.org/" + "wikipedia/commons/7/7d/Dog_face.png")
image = imread(fic_exemple, as_gray = True)
plt.imshow(image, cmap = cm.gray)
plt.show()
```
